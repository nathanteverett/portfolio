# 10-Year Mean-Reversion Investment Strategy Backtest
By Nathan T. Everett

## Introduction

Below is the implementation of a mean-reversion trading strategy based on the simple moving average (SMA) of top performing stocks. The data comes from yfinance, a python package that utilizes Yahoo's API on historical market data. The data set provides daily stock prices. Therefore, all of the prices are evaluated daily on market close.

The goal of this strategy is to get more exposure to the top companies of the S&P 500 than the index or its ETFs alone. Of course being so closely related to the S&P will mean that it will be strongly correlated. However, the goal of this strategy is to investigate if exposure to the top 10 companies alone will be significant enough to outperform the index as a whole. 

## Methods

In the following backtest, five 10-year periods were selected: 2015-2025, 2011-2021, 2005-2015, 2000-2010, 1995-2005. These periods were chosen to investigate the effect in bull markets like 2011-2021 and 2015-2025, and bear markets where a bubble popped in the middle and end of these periods, like 1995-2005 and 2000-2010, respectively. 

The strategy looks at the top 10 performing stocks in the S&P 500 at the beginning of the starting date. When a stock dips below its 20-day SMA by 10%, the stock is purchased. The amount of cash spent of each stock is fixed at \\$500 per investment. In addition to a modest starting amount of \\$10,000, a cash flow of \\$1,000 per month is also assumed for the duration. Furthermore, a brokerage fee of 2% annually, compounding annually, is assumed. No selling of stocks occurs. 

For each period, the portfolio was compared to a  simple dollar cost averaging (DCA) strategy on the SPY ETF. The initial amount was put into the SPY right away, and each monthly cash flow was immediately invested. 

## Results

For strong bull markets, this strategy works well. Most noticably this strategy outperformed the S&P in the 2015-2025 period. In this era, the strategy yielded 243.72% ROI while the SPY portfolio yielded 123.10% ROI.

However, all of the other four periods did not out-perform the SPY. This likely happened for a few reasons. Firstly, many of these stocks are not mean-reverting; many companies that are in the top of the S&P at the start time did not stay in the top by the end. While the S&P restructures often, this strategy follows the companies until the end of the period. 

Furthermore, this strategy was very weak in bear markets. In the worst case, the strategy yielded a negative ROI of -8.41% in the 2000-2010 period. The SPY portfolio didn't do well at this time either, as the market was in its early recovery phase after the 2008 financial crisis, but the losses were still more pronounced than compared to the SPY portfolio.

## Conclusion

The strategy is not effective in its current state. In the era of the magnificent 7 it would likely perform very well. However, a bull run that we have seen in the last 15 years is unprecedented. There would be no way to guess that this strategy would work without forward knowledge. The S&P wins again.

## Outlook 

There are many things that could improve this strategy. Firstly, the portfolio should be more adaptive. Stocks should be sold at relative peaks and there should be preparation for restructuring to reflect the S&P. Secondly, the threshold for the SMA should correlate with the amount of stock purchased. The better the deal of the stock, the more you should buy. Finally, SMA average time should be optimized. Many of the parameters in this investigation were kept constant. Adaptive parameters could help yield higher returns and protect against non-mean-reverting stocks

## Main Program

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from yahoofinancials import YahooFinancials
import pandas as pd
import datetime

sns.set_theme()
```

```python
# Create SMA function
def sma(asset_df, avg_time):
    return asset_df.rolling(window=avg_time).mean()
```

### Period 1: 2015-2025

```python
# Period 1: 2015:2025
"""
Top 10 stocks in S&P in 2015:

    Apple Inc.
    Exxon Mobil Corporation
    Microsoft Corporation
    Berkshire Hathaway Inc.
    Alphabet Inc. (Google)
    Johnson & Johnson
    Wells Fargo & Company
    General Electric Company
    Procter & Gamble Company
    JPMorgan Chase & Co.
    """
symbols = ['AAPL','XOM', 'MSFT', 'BRK-B', 'GOOG', 'JNJ', 'WFC', 'GE', 'PG', 'JPM']

stocks_df = pd.DataFrame()
for symbol in symbols:
    stock = yf.download([symbol],start='2015-01-01', end='2025-01-01',progress=False)
    stocks_df[symbol] = stock['Close']

spy = yf.download('SPY', start='2015-01-01', end='2025-01-01',progress=False)
stocks_df.index = stocks_df.index.date
spy.index = spy.index.date
spy = spy['Close'].squeeze()

stocks_df.head()
```

```python
# Demonstrate SMA function

sma_df = sma(stocks_df['AAPL'].iloc[-120:-1], 20) #20 day SMA
sma_df.plot(label = '20-Day SMA')
plt.title('Apple 20-Day SMA and Stock Price')
stocks_df['AAPL'].iloc[-100:-1].plot(label='AAPL Stock Price')
plt.xlabel('Date')
plt.ylabel('Price ($)')
```

```python
# SMA 2015:2025 period

good_deal = stocks_df < sma(stocks_df,20)*.90 #make mask of stocks that are below threshold
portfolio = dict.fromkeys(symbols, 0) #initialize empty portfolio
prev_month = good_deal.index[0].month #tracking month for cashflow
prev_year = good_deal.index[0].year #tracking year for brokerage fees
profit_tracker = dict.fromkeys(symbols, 0) 

# Init params
cash = 10000
monthly_cashflow = 1000
total_invest = 0
total_cash = cash
invest_amt = 500
fee = 0.02 # 2% annual fee


# Compare to SPY
"""Comparing investment strategy to putting every dollar in SPY as the cashflow comes in."""
portfolio_spy = {"SPY": cash/spy.iloc[0]}

# Store tracked profits
profit_df = pd.DataFrame(columns = ['Portfolio', 'SPY Portfolio'], index = stocks_df.index, 
                         data = np.zeros((len(stocks_df.index), 2)))


for date, row in good_deal.iterrows(): #iter through df, need to iterate to do sequentially

    month_date = date.month
    year_date = date.year
    
    # Cash flow
    if month_date != prev_month:
        prev_month = month_date
        cash += monthly_cashflow
        total_cash += monthly_cashflow
        portfolio_spy['SPY'] += monthly_cashflow/spy.loc[date] #buy SPY
        
    # Annual account fee
    if year_date != prev_year:
        prev_year = year_date
        cash -= fee*sum(portfolio.values()) # 2% of AUM
    
    if row.any():      #what rows with a true
        true_columns = row[row == True].index.tolist() #get true columns 

        # Buy stocks if good deal
        for col in true_columns:  #Go through each stock that is a deal on this day
            if cash >= invest_amt: #Make sure you have enough cash
                portfolio[col] += invest_amt/stocks_df.loc[date, col] #buy stock
                cash -= invest_amt  #adjust cash
                total_invest += invest_amt #track investment

    # Track profits
    for key in portfolio:
        profit_tracker[key] = portfolio[key]*stocks_df[key][date]

    profit_df['Portfolio'].loc[date] = sum(profit_tracker.values()) - total_invest
    profit_df['SPY Portfolio'].loc[date] = sum(portfolio_spy.values())*spy[date] - total_cash
    


profit_df['Portfolio ROI'] = round(profit_df['Portfolio']/total_invest*100, 2) 
profit_df['SPY ROI'] = round(profit_df['SPY Portfolio']/total_cash*100,2)   


print("Results for 2015-2025:",
      '\n--------------------------------------------'
    "\n10-year portfolio profit:", round(profit_df['Portfolio'].iloc[-1], 2),
      "\n10-year portfolio ROI:", round(profit_df['Portfolio ROI'].iloc[-1],2),"%", 
      "\nleftover cash:", round(cash,2),'\n')

print("10-year SPY DCA profit:", round(profit_df['SPY Portfolio'].iloc[-1],2), 
      "\n10-year portfolio ROI:", round(profit_df['SPY ROI'].iloc[-1],2),"%")
```

```python
# Create the plot
fig, ax1 = plt.subplots()

# Plot the original returns on the first y-axis
ax1.plot(profit_df.index, profit_df['Portfolio'], 'r', label='Porfolio')
ax1.set_xlabel('Date')
ax1.set_ylabel('Portfolio Value', color='black')
ax1.tick_params( labelcolor='black',labelrotation=45)

# Create a second y-axis for percent returns
ax2 = ax1.twinx()
ax2.set_ylabel(' ROI (%)', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Plot SPY Portfolio
ax1.plot(profit_df.index, profit_df['SPY Portfolio'], 'b', label='SPY Porfolio')

# Ensure scaling on ROI axis
ax2.set_ylim(profit_df['Portfolio ROI'].min(), profit_df['Portfolio ROI'].max())

# Add legends
ax1.legend(loc='upper left')

# Turn second grid off
ax2.grid(False)

# Show the plot
plt.title('Trading Strategy Portfolio vs. S&P 500 DCA (2015-2025)')
```

```python
# SMA 1995:2005 period

good_deal = stocks_df < sma(stocks_df,20)*.90 #make mask of stocks that are below threshold
portfolio = dict.fromkeys(symbols, 0) #initialize empty portfolio
prev_month = good_deal.index[0].month #tracking month for cashflow
prev_year = good_deal.index[0].year #tracking year for brokerage fees
profit_tracker = dict.fromkeys(symbols, 0) 

# Init params
cash = 10000
monthly_cashflow = 1000
total_invest = 0
total_cash = cash
invest_amt = 500
fee = 0.02 # 2% annual fee


# Compare to SPY
"""Comparing investment strategy to putting every dollar in SPY as the cashflow comes in."""
portfolio_spy = {"SPY": cash/spy.iloc[0]}

# Store tracked profits
profit_df = pd.DataFrame(columns = ['Portfolio', 'SPY Portfolio'], index = stocks_df.index, 
                         data = np.zeros((len(stocks_df.index), 2)))


for date, row in good_deal.iterrows(): #iter through df, need to iterate to do sequentially

    month_date = date.month
    year_date = date.year
    
    # Cash flow
    if month_date != prev_month:
        prev_month = month_date
        cash += monthly_cashflow
        total_cash += monthly_cashflow
        portfolio_spy['SPY'] += monthly_cashflow/spy.loc[date] #buy SPY
    
    # Annual account fee
    if year_date != prev_year:
        prev_year = year_date
        cash -= fee*sum(portfolio.values()) # 2% of AUM
        
    if row.any():      #what rows with a true
        true_columns = row[row == True].index.tolist() #get true columns 

        # Buy stocks if good deal
        for col in true_columns:  #Go through each stock that is a deal on this day
            if cash >= invest_amt: #Make sure you have enough cash
                portfolio[col] += invest_amt/stocks_df.loc[date, col] #buy stock
                cash -= invest_amt  #adjust cash
                total_invest += invest_amt #track investment
    
    # Track profits
    for key in portfolio:
        profit_tracker[key] = portfolio[key]*stocks_df[key][date]

    profit_df['Portfolio'].loc[date] = sum(profit_tracker.values()) - total_invest
    profit_df['SPY Portfolio'].loc[date] = sum(portfolio_spy.values())*spy[date] - total_cash
    


profit_df['Portfolio ROI'] = round(profit_df['Portfolio']/total_invest*100, 2) 
profit_df['SPY ROI'] = round(profit_df['SPY Portfolio']/total_cash*100,2)   


print("Results for 1995-2005:",
      '\n--------------------------------------------'
    "\n10-year portfolio profit:", round(profit_df['Portfolio'].iloc[-1], 2),
      "\n10-year portfolio ROI:", round(profit_df['Portfolio ROI'].iloc[-1],2),"%", 
      "\nleftover cash:", round(cash,2),'\n')

print("10-year SPY DCA profit:", round(profit_df['SPY Portfolio'].iloc[-1],2), 
      "\n10-year portfolio ROI:", round(profit_df['SPY ROI'].iloc[-1],2),"%")
```

```python
# Create the plot
fig, ax1 = plt.subplots()

# Plot the original returns on the first y-axis
ax1.plot(profit_df.index, profit_df['Portfolio'], 'r', label='Porfolio')
ax1.set_xlabel('Date')
ax1.set_ylabel('Portfolio Value', color='black')
ax1.tick_params( labelcolor='black',labelrotation=45)

# Create a second y-axis for percent returns
ax2 = ax1.twinx()
ax2.set_ylabel(' ROI (%)', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Plot SPY Portfolio
ax1.plot(profit_df.index, profit_df['SPY Portfolio'], 'b', label='SPY Porfolio')

# Ensure scaling on ROI axis
ax2.set_ylim(profit_df['Portfolio ROI'].min(), profit_df['Portfolio ROI'].max())

# Add legends
ax1.legend(loc='upper left')

# Turn second grid off
ax2.grid(False)

# Show the plot
plt.title('Trading Strategy Portfolio vs. S&P 500 DCA (1995-2005)')
```
